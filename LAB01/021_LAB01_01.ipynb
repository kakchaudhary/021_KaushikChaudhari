```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#1) Draw Scatter Plot between age and salary for "Data_for_Transformation.csv" file
data=pd.read_csv('Data_for_Transformation.csv')
print(data.head())
plt.scatter(data['Age'],data['Salary'])
plt.show()
```

```python
#2) Draw Histogram of column/feature "Salary"
plt.hist(data['Salary'],bins=5)
plt.show()
```

```python
#3) Plot pie chart for column/feature "Country"
figure = plt.figure()
ax = figure.add_axes([0,0,1,1])
ax.axis('equal')
country = data['Country']
sal = data['Salary']
ax.pie(sal,labels=country,autopct='%1.2f%%')
plt.show()
```
